## Check a number in array

```python
def checkNumber(arr, x):
    # Please add your code here
    #print(arr[0])
    if len(arr)==0:
        return False

    if arr[0]==x:
        return True

    arr1=arr[1:]
    return checkNumber(arr1, x)

from sys import setrecursionlimit
setrecursionlimit(11000)

n=int(input())
arr=list(int(i) for i in input().strip().split(' '))
x=int(input())
if checkNumber(arr, x):
    print('true')
else:
    print('false')
```

## Factorial

```python
def factorial_of_n(n):
    if n==1:
        return n
    return n*factorial_of_n(n-1)
n=4
factorial_of_n(n)
```

## Fibonacci

```python
def fibonacci(n):
  if n==1 or n==2:
    return 1
  return fibonacci(n-1)+fibonacci(n-2)
n=3  #0,1,1,2,3,5,..

fibonacci(n)
```
